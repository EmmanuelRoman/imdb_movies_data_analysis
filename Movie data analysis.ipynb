## Data Preparation
There are some columns that will not be helpful for our analysis. Lets remove them with the .drop method:    
   
   - `movie_imdb_link`
   - `num_critic_for_reviews`
   - `genre`

```python
df.drop(axis=1, columns=['movie_imdb_link', 'num_critic_for_reviews', 'genres'], inplace=True)
df.head()
```

Let's check the shape again in order to be sure we did removed those columns correctly:

```python
df.shape
```

Let's convert the dtype of `title_year` column to integer

```python
df['title_year']=df['title_year'].astype('Int64')
df.head()
```

Let's check the type again:

```python
df['title_year'].dtype
```

Renaming some columns, using a dictionary: 
- changing `gross` to `movie_income`
-          `language` to `Language`
-           `budget` column to `movie_budget`

```python
#We create the dict:
rename_dict = {'gross':'movie_income', 'language':'Language', 'budget':'movie_budget'}
```

```python
df.rename(columns=rename_dict, inplace=True )
df.head()
```

### Data Analysis 


Let's checkout the distribution of `duration` column using a histogram.


```python
hist_duration=df.duration.hist(bins=35, grid = False, figsize=(10,8))
hist_duration.set_xlabel('Movie duration')
hist_duration.set_ylabel('Count')
hist_duration.set_title('Movie duration histogram')
```

Lets create a hitogram to see the distribution of `imdb_score` column

```python
imdb_score_hist=df.imdb_score.hist(bins=25, grid=False, figsize=(10,8))
imdb_score_hist.set_xlabel('IMDB Score')
imdb_score_hist.set_ylabel('Count')
imdb_score_hist.set_title('IMDB Score hist')
```

Let's check how many movies have the `imdb_score` less than `4`.

```python
low_score_movies=df[df.imdb_score < 4]
low_score_movies.head()
```

### Let's answer to some analytical questions 
    
Question 1 - In which year we have the `highest` and `lowest` number of movies?

Now finding out number of movies per year

```python
movies_per_year = df.groupby('title_year').size()
movies_per_year
```

Let's create a line chart to show the result

```python
movies_per_year.plot()
```

Question 2 - In which year we have `minimum` and `maximum` average `imdb_score` ? 

Finding average imdb score per year 

```python
avg_scores=df.groupby('title_year')['imdb_score'].mean()
avg_scores
```

Let's Use a barchart to visualize the result

```python
avg_barchart=avg_scores.plot.bar(figsize=(10,8), title='Average imdb score per year')
avg_barchart.set_ylabel('imdb Score')
```

Exercise: 
Question 3 - In which year we have the `minimum` and `maximum` spent movie budget?

`hint: use a line chart to visualize your results.` 

```python
chart = df.groupby('title_year')['movie_budget'].sum().plot()
chart.set_title('Movies budget per year')
chart.set_ylabel('Budget')
```

It looks like we got the maximun value after 2000. Lets verify that year and its budget:

```python

print('Our minimun spent movies budget was', df.groupby('title_year')['movie_budget'].sum().min())
print('And its year was', df.groupby('title_year')['movie_budget'].sum().idxmin())

print ('')
print('Our maximun spent movies budget was', df.groupby('title_year')['movie_budget'].sum().max())
print('And its year was', df.groupby('title_year')['movie_budget'].sum().idxmax())

```

Question 4 - Is there any relationship between IMdb score of a movie and the movie income?

To approaches to solve this question:
- Visual Correlation Analysis 
- Generating the Correlation Matirx

```python
#First, we'll create a df with the values we're interested in
data = df[['imdb_score','movie_income']]
data
```

Let's drop the rows with `non` values

```python
data = data.dropna()
data
```

Let's create a scatter plot to see the data

```python
scatter_plot = data.plot.scatter(x='imdb_score', y='movie_income', figsize=(7,7))
scatter_plot
```

Since its not clear at all the existence of a correlation between those two metrics. Let's generate the correlation matrix

```python
data.corr()
```

According to our data, the correlation value between these two metrics is about 20%. We can then conclude that actually there's no a strong correlation at all between a movie imdb score and its income
