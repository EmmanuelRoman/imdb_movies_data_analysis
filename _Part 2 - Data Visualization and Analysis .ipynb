Answer to the following Analytical questions: 

Question 1: Which country has gained the highest average IMdb Score according to our data? <br>
Use a bar chart to visualize your results

```python
#First, we filter our dataframe and plot it
d1=df.groupby('country').imdb_score.mean()
plot1=d1.plot.bar(figsize=(10,8))
plot1.set_ylabel('Average imdb score')
plot1.set_title('Average imdb score per country')
#Just to verify 
df.groupby('country').imdb_score.mean().idxmax()

```
